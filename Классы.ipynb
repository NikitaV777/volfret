```python
class Vehicle:
    def __init__(self, maximum_speed, year, divergence, volume_of_transportation):
        self._maximum_speed = maximum_speed
        self.__year = year
        self._divergence = divergence
        self._volume_of_transportation = volume_of_transportation
        
       

    def iMaximum_speed(self, maximum_speed):
        self._maximum_speed = maximum_speed

    def iYear(self, year):
        self.__year = year

    def iDivergence(self, divergence):
        self._divergence = divergence

    def iVolume_of_transportation(self, volume_of_transportation):
         self._volume_of_transportation = volume_of_transportation

    def getMaximum_speed(self):
        return self._maximum_speed

    def getYear(self):
        return self.__year

    def getDivergence(self):
        return self.__divergence

    def getVolume_of_transportation(self):
        return self.__volume_of_transportation


    def Display_Vehicle(self):
        print('Inventory unit: %s' % self.__class__.name)
        print('Maximum_speed: ' , self.__maximum_speed)
        print('Year: ', self.__year)
        print('Divergence: ' , self.__divergence)
        print('Volume of transportation: ', self.__volume_of_transportation)

class Car(Vehicle):
    name = 'Car'
    def __init__(self, maximum_speed, year, divergence, volume_of_transportation,x):
        Vehicle.__init__(self, maximum_speed, year, divergence, volume_of_transportation)
        self.__doors  = x
    def iDoors(self, doors):  
        self.__doors = doors
    def getDoors(self):
        return self.__doors
    def Display(self):
        self.Display_Vehicle()
        print('Number of doors: ', self.__doors)

class Truck(Vehicle):
    name = 'Truck'
    def __init__(self, maximum_speed, year, divergence, volume_of_transportation,x):
        Vehicle.__init__(self, maximum_speed, year, divergence, volume_of_transportation)
        self.__drive  = x
    def iDrive(self, drive):  
        self.__drive = drive
    def getDrive(self):
        return self.__drive
    def Display(self):
        self.Display_Vehicle()
        print('Drive type: ', self.__drive)

class Half_trucks(Vehicle):
    name = 'Half_trucks'
    def __init__(self, maximum_speed, year, divergence, volume_of_transportation,x):
        Vehicle.__init__(self, maximum_speed, year, divergence, volume_of_transportation)
        self.__passengers = x
    def iCapacity(self, passengers):
        self.__passengers = passengers
    def getCapacity(self):
        return self.__passengers
    def Display(self):
        self.Display_Vehicle()
        print('Number of passengers: ', self.__passengers)

def main():
    inventory = []
    while True:
        classType = input('Is the vehicle a car, truck, or half trucks?  ')
        print (classType)
        maximum_speed = input('Please enter the maximum_speed  of the %s: ' % classType)
        year = input('Please enter the year of the %s: ' % classType)
        divergence  = input('Please enter the divergence of the %s: ' % classType)
        volume_of_transportation = input('Please enter the volume_of_transportation of the %s: ' % classType)
        if classType == 'car':        
            x = input('Please enter the amount of doors on the car: ')
            inventory.append(Car(maximum_speed, year, divergence, volume_of_transportation, x))
        elif classType == 'truck':
            x = input('Please enter 2 wheel or 4 wheel drive for the truck: ')
            inventory.append(Truck(maximum_speed, year, divergence, volume_of_transportation, x))
        elif classType == 'half trucks':
            x = input('Please enter the capacity of the half trucks: ')
            inventory.append(Half_trucks(maximum_speed, year, divergence, volume_of_transportation,x))
        print('\n\n')
        inventory[-1].Display()
        cont = 'go'
        while cont not in ('y','n'):
            cont = input('Would you like to add another vehicle?  y/n  ')
        if cont == 'n':
            print( inventory )
            break

if __name__ == '__main__':
    main()
```
